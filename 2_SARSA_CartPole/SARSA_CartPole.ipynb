# SARSA with Cart Pole environment

## 1. Import necessary libraries

```python
# import libraries
import gym
import numpy as np
import time
import matplotlib.pyplot as plt
import csv
```

## 2. Implement the SARSA Algorithms

```python
class SARSA: 
    def __init__(self, env, alpha, gamma, epsilon, numberEpisodes, numberOfBins, lowerBounds, upperBounds):
        import numpy as np
        self.env = env  # Cart Pole environment
        self.alpha = alpha  # step size
        self.gamma = gamma  # discount factor
        self.epsilon = epsilon  # parameter for epsilon-greedy approach
        self.actionNumber = env.action_space.n  # enumerate actions
        self.numberEpisodes = numberEpisodes  # total number of simulation episodes
        self.numberOfBins = numberOfBins  # number of discrete points

        # Limits for state discretization [cart position, cart velocity, pole angle, and pole angular velocity]
        self.lowerBounds = lowerBounds  # lower bound [1, 0, 2, 3]
        self.upperBounds = upperBounds  # upper bound

        # Sum of rewards in every learning episode (list)
        self.sumRewardsEpisode = []

        # Action value function matrix
        self.Qmatrix = np.random.uniform(low=0, high=1, size=(
            numberOfBins[0], numberOfBins[1], numberOfBins[2], numberOfBins[3], self.actionNumber))
 
    def returnIndexState(self, state):
        position = state[0] #position will at the position 0 of the state list
        velocity = state[1] #velocity will be at the position 1 of the state list
        angle = state[2]    #angle will be at the position 2 of the state list
        angularVelocity = state[3] #angular velocity will be at the position 3 of the state list

        # Evenly spaced numbers over a specified range to discretization the state parameters
        cartPositionBin = np.linspace(self.lowerBounds[0], self.upperBounds[0], self.numberOfBins[0])
        cartVelocityBin = np.linspace(self.lowerBounds[1], self.upperBounds[1], self.numberOfBins[1])
        poleAngleBin = np.linspace(self.lowerBounds[2], self.upperBounds[2], self.numberOfBins[2])
        poleAngleVelocityBin = np.linspace(self.lowerBounds[3], self.upperBounds[3], self.numberOfBins[3])

        # Enumerate the discrete states
        indexPosition = np.maximum(np.digitize(state[0], cartPositionBin) - 1, 0)
        indexVelocity = np.maximum(np.digitize(state[1], cartVelocityBin) - 1, 0)
        indexAngle = np.maximum(np.digitize(state[2], poleAngleBin) - 1, 0)
        indexAngularVelocity = np.maximum(np.digitize(state[3], poleAngleVelocityBin) - 1, 0)
        
        return tuple([indexPosition, indexVelocity, indexAngle, indexAngularVelocity])


    def selectAction(self,state,index):
         
        # first 100 episodes we select completely random actions to avoid being stuck
        if index<100:
            return np.random.choice(self.actionNumber)   
             
        # Returns a random real number in the half-open interval [0.0, 1.0)
        randomNumber=np.random.random()
           
        if index>1000:
            self.epsilon=0.9*self.epsilon
         
        # if this condition is satisfied, we are exploring, that is, we select random actions
        if randomNumber < self.epsilon:
            # returns a random action selected from: 0,1,...,actionNumber-1
            return np.random.choice(self.actionNumber)            
         
        # otherwise, we are selecting greedy actions
        else:
            # we return the index where actionValueMatrixEstimate[state,:] has the max value
            return np.random.choice(np.where(self.Qmatrix[self.returnIndexState(state)] == np.max(self.Qmatrix[self.returnIndexState(state)]))[0])
            # 0 at the end since np.where returns a tuple, choose first action

    def simulateEpisodes(self): # Simulating learning episodes
        import numpy as np

        # Loop through episodes
        for indexEpisode in range(self.numberEpisodes):

            # Record rewards in each episode
            rewardsEpisode = []

            # reset the environment at the beginning of every episode
            (stateS, _) = self.env.reset()
            stateS = list(stateS)

            print("Simulating episode {} \n".format(indexEpisode))

 
            # here we step from one state to another until a terminal state is reached
            terminalState = False
            actionA = self.selectAction(stateS, indexEpisode)  # Initial action selection

            while not terminalState:

                # Return a discretized index of the state
                stateSIndex = self.returnIndexState(stateS)
                
                #Here, reward is number of timesteps it lasts
                (stateSnext, reward, terminalState, _, _) = self.env.step(actionA)
                rewardsEpisode.append(reward)
                stateSnext = list(stateSnext)
                stateSnextIndex = self.returnIndexState(stateSnext)
            
                # Select action for the next state
                actionAnext = self.selectAction(stateSnext, indexEpisode)
            
                if not terminalState:
                    actionAnext = self.selectAction(stateSnext, indexEpisode)
                    # stateS+(actionA,) - ppend the tuples
                    # for example, for stateS=(0,0,0,1) and actionA=(1,0)
                    # we have stateS+(actionA,)=(0,0,0,1,0)
                    error = reward + self.gamma * self.Qmatrix[stateSnextIndex + (actionAnext,)] - self.Qmatrix[stateSIndex + (actionA,)]
                else:
                    # in the terminal state, we have Qmatrix[stateSnext,actionAnext]=0 (Q'=0)
                    error = reward - self.Qmatrix[stateSIndex + (actionA,)]
                    print("Sum of rewards {}".format(np.sum(rewardsEpisode)))
                
                # Update Q-value
                self.Qmatrix[stateSIndex + (actionA,)] += self.alpha * error
            
                # Update current state and action for the next step
                stateS = stateSnext
                actionA = actionAnext

            
            
            self.sumRewardsEpisode.append(np.sum(rewardsEpisode))

   
    # simulate the final learned optimal policy

    def simulateLearnedStrategy(self):
        import gym
        import time
        env1 = gym.make('CartPole-v1', render_mode='human')
        (currentState, _) = env1.reset()
        env1.render()
        timeSteps = 1000
        # obtained rewards at every time step
        obtainedRewards = [] # list of obtained rewards during time steps of a single episode

        for timeIndex in range(timeSteps):
            #print(timeIndex) ###################################### đang xóa timeIndex
            # select greedy actions
            actionInStateS = np.random.choice(np.where(self.Qmatrix[self.returnIndexState(currentState)] == np.max(
                self.Qmatrix[self.returnIndexState(currentState)]))[0])
            #Choose greedy
            currentState, reward, terminated, truncated, info = env1.step(actionInStateS)
            obtainedRewards.append(reward)
#             time.sleep(0.05) vừa xóa dòng này
            if (terminated):
#                 time.sleep(1) vừa xóa dòng này để train cho nhanh
                break
        return obtainedRewards, env1

    
    #    function for simulating random actions many times
    #   this is used to evaluate the optimal policy and to compare it with a random policy
    #  OUTPUT:
    # sumRewardsEpisodes - every entry of this list is a sum of rewards obtained by simulating the corresponding episode

    def simulateRandomStrategy(self):
        import gym
        import time
        import numpy as np
        env2 = gym.make('CartPole-v1', render_mode='human')
        (currentState, _) = env2.reset()
        env2.render()
        # number of simulation episodes
        episodeNumber = 100
        # time steps in every episode
        timeSteps = 1000
        # sum of rewards in each episode
        sumRewardsEpisodes = []

        for episodeIndex in range(episodeNumber):
            rewardsSingleEpisode = []
            initial_state = env2.reset()
            print(episodeIndex)
            for timeIndex in range(timeSteps):
                random_action = env2.action_space.sample()
                observation, reward, terminated, truncated, info = env2.step(random_action)
                rewardsSingleEpisode.append(reward)
                if terminated:
                    break
            sumRewardsEpisodes.append(np.sum(rewardsSingleEpisode))
        return sumRewardsEpisodes, env2
```

## 4. Define the necessary parameters

```python
# Define the parameters for state discretization
upperBounds = env.observation_space.high
lowerBounds = env.observation_space.low
cartVelocityMin = -3
cartVelocityMax = 3
poleAngleVelocityMin = -10
poleAngleVelocityMax = 10
upperBounds[1] = cartVelocityMax
upperBounds[3] = poleAngleVelocityMax
lowerBounds[1] = cartVelocityMin
lowerBounds[3] = poleAngleVelocityMin

#Discretization regions (evenly spaced regions to enumerate the states)
numberOfBinsPosition = 30
numberOfBinsVelocity = 30
numberOfBinsAngle = 30
numberOfBinsAngleVelocity = 30
numberOfBins = [numberOfBinsPosition, numberOfBinsVelocity, numberOfBinsAngle, numberOfBinsAngleVelocity]

numberEpisodes = 20000
```

## 5. Simulate episode and training & exporting the data file

```python
# Create objects for three runs (env, alpha, gamma, epsilon, numberEpisodes, numberOfBins, lowerBounds, upperBounds)
Q1 = SARSA(env, 0.1, 0.9, 0.2, numberEpisodes, numberOfBins, lowerBounds, upperBounds)
Q2 = SARSA(env, 0.08, 1.02, 0.15, numberEpisodes, numberOfBins, lowerBounds, upperBounds)
Q3 = SARSA(env, 0.075, 1.1, 0.1, numberEpisodes, numberOfBins, lowerBounds, upperBounds)

# Run the algorithm for each object
# Q1.simulateEpisodes()
# Q2.simulateEpisodes()
# Q3.simulateEpisodes()
```

```python
#Export all rewards to a CSV file

def export_rewards_to_csv(filename='rewards.csv'):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Episode', 'Run 1', 'Run 2', 'Run 3'])

        # Running the Q-learning algorithm three times with three different seeds
        for idx in range(numberEpisodes):
            writer.writerow([idx + 1, Q1.sumRewardsEpisode[idx], Q2.sumRewardsEpisode[idx], Q3.sumRewardsEpisode[idx]])
    print(f"Rewards exported to {filename}")
    
# # Combine the sum of rewards from all runs into a single list
# sumRewardsCombined = [Q1.sumRewardsEpisode, Q2.sumRewardsEpisode, Q3.sumRewardsEpisode]

# Export the rewards to a CSV file
export_rewards_to_csv()
```
